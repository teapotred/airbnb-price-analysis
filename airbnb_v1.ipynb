```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('airbnb_listings.csv')
df
```

```python
df=df[df['price'].notna()&(df['price']!='')]
df
```

```python
df=df.drop(columns=['description','neighborhood_overview','last_scraped','first_review','last_review','host_since','calendar_last_scraped','host_location','host_location','host_about','host_verifications','host_has_profile_pic','host_identity_verified','calendar_updated','instant_bookable','scrape_id','listing_url','picture_url','host_url','host_thumbnail_url','host_picture_url','host_is_superhost','host_response_rate','host_acceptance_rate','host_neighbourhood'])
df
```

```python
df = df.drop(columns=['source','host_response_time','host_name','neighbourhood'])
df
```

```python
df
```

```python
df['price'] = df['price'].replace('[$,]', '', regex=True).astype(float)
df
```

```python
price_df = df['price']
price_df
```

```python
df['price']=pd.to_numeric(df['price'])
df
```

```python
df['price'].describe()
```

```python

df['amenities'] = df['amenities'].str.replace(r'[{}"]', '', regex=True)
df['amenities_list'] = df['amenities'].str.split(',')

amenities_to_check = ['Wifi', 'Kitchen', 'Air conditioning', 'Heating', 'Washer', 'Dryer']

for amenity in amenities_to_check:
    df[amenity.strip().replace(" ", "_").lower()] = df['amenities_list'].apply(lambda x: 1 if amenity in [a.strip() for a in x] else 0)
```

```python
df['wifi'].describe()
```

```python
df.head()
```

```python
df['price'] = df['price'].replace('[$,]','',regex=True).astype(float)
```

```python
##plot roomtype vs avg price 

avg_price=df.groupby('room_type')['price'].mean().sort_values(ascending=False)

avg_price.plot(kind='bar', color ='skyblue', edgecolor='black')
plt.title('Room type vs Average Price')
plt.ylabel('Average price ($)')
plt.xlabel('Room type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

custom_bins = np.arange(0, 2100, 100)  # From 0 to 2000 in $100 increments


plt.figure(figsize=(10,6))
plt.hist(df['price'], bins=custom_bins, color='yellowgreen', edgecolor='black')
plt.title('Airbnb Price Distribution ($100 Bins)')
plt.xlabel('Price ($)')
plt.ylabel('Number of Listings')
plt.xticks(custom_bins, rotation=45)  
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

top_neighbourhoods = df['neighbourhood_cleansed'].value_counts().nlargest(10).index
df_top =df[df['neighbourhood_cleansed'].isin(top_neighbourhoods)]


pivot_table = df_top.pivot_table(
    values='price',
    index='neighbourhood_cleansed',          
    columns='room_type',
    aggfunc='mean'
).round(0)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    pivot_table,            
    fmt=".0f",               
    cmap='YlOrRd',          
    linewidths=0.5,
    linecolor='gray'
)

plt.title('Average Airbnb Price by City and Room Type')
plt.xlabel('Room Type')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```
